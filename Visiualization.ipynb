```python
import matplotlib.pyplot as plt
import numpy as np

names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,83,77,78,95]
age=[20,45,30,35,50]

GradeList = zip(names,grades,age)
df = pd.DataFrame(data = GradeList, columns=['Names', 'Grades','Age'])

x = np.arange(len(df.Names))  # Ly v tr ca cc gi tr trong trng Names
width = 0.35  #  rng ca thanh biu 
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df.Grades, width, label='Grades')
rects2 = ax.bar(x + width/2, df.Age, width, label='Age')

ax.set_ylabel('Y')
ax.set_xlabel('Name')
ax.set_title('Biu  gp nhm theo tn ca trng Grades v Age')
ax.set_xticks(x)
ax.set_xticklabels(df.Names)
ax.legend()
```

# 2. Box Plot

Biu  dng hp Box Plot

Biu  hp (Box-plot) l biu  din t 5 v tr phn b ca d liu,  l : 
* gi tr nh nht (min), 
* t phn v th nht (Q1), 
* trung v (median), 
* t phn v th 3 (Q3) v 
* gi tr ln nht (max).


### S dng hm boxplot()
Biu  vi 1 trng d liu.


```python
import matplotlib.pyplot as plt
import pandas as pd

names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
age=[20,45,30,35,50]
gender = ['Male','Female','Female','Male','Female']
status = ['Senior','Senior','Junior','Junior','Senior']

GradeList = zip(names,grades,gender,age,status)
df = pd.DataFrame(data = GradeList, columns=['Names', 'Grades', 'Gender','Age', 'Status'])
df.boxplot(column='Grades')
```

### V biu  vi 2 trng d liu Grades, Age v t mu phn Box

```python
df.boxplot(column=['Grades', 'Age'],patch_artist=True)
# patch_artist=True: t mu phn Box
```

### V biu  trn trng Grades gp nhm trn trng Gender

```python
df.boxplot(by='Gender', column='Grades')
```

# 3. Biu  Histogram
* Histogram l mt dng biu  th hin phn b d liu,m t d liu mt cch n gin m khng lm mt bt c thng tin thng k no ca d liu.


###  ngha biu  Histogram
* Histogram cho thy hnh thi phn b ca d liu qua  thit lp mc tiu v xu hng khc phc cho tng vn . Biu  tr li cc cu hi sau:

    * Kiu phn b d liu?
    *  rng d liu nh th no?
    * D liu c i xng hay khng?
    * C d liu no nm ngoi hay khng?
    
### i tng ap dung cua Histogram
* D liu l s
* Khi mun xem hnh dng ca phn phi d liu, c bit l khi xc nh xem u ra ca mt quy trnh c c phn phi bnh thng khng
* Phn tch xem mt quy trnh c th p ng cc yu cu ca khch hng khng
* Phn tch kt qu u ra t quy trnh ca nh cung cp trng nh th no
* Xc nh xem u ra ca hai hay nhiu qu trnh c khc nhau khng
* p dng cho cc i tng c sn lng d liu ln.

### S dng hm nh sau:
* .hist(x, bins=10, range=None, normed=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, hold=None, data=None, **kwargs)

#### Hai tham s quan trng  l x v bins:
* x l d liu dng  v biu 
* bins l s ngn  cho Python bit bao nhiu ngn m d liu s c chia ra. Da trn s lng ny, hist() s t ng tm ra ranh gii thch hp cho cc ngn v tnh ton c bao nhiu gi tr tng ng vi  ngn ; nu ta khng c t tham s ny th mc nh n l 10.


```python
# V d: V biu  th hin s phn b ca trng hours trong file d liu gradedata.csv vi s bin =6.
import matplotlib.pyplot as plt
import pandas as pd
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()

#v biu 
df.hist(column="hours",color='blue', bins=6)
#hoc c th dng lnh sau nhng khng c tiu 
plt.hist(df.hours, color='blue', bins=6)
```

# 4. Biu  hnh trn (Pie Chart)

### .pie(x, explode=None, labels=None, colors=None, autopct=None, pctdistance=0.6, shadow=False, labeldistance=1.1, startangle=None, radius=None, counterclock=True, wedgeprops=None, textprops=None, center=(0, 0), frame=False, rotatelabels=False, \*, data=None)

### Gii thiu mt s tham s hay s dng
* x: l trng d liu  v biu 
* explode: Mc nh l none, cn nu c gi tr ca phn no th phn biu   s lch ra hoc c bn knh ln hn nhng phn cn li.
* colors: mu ca cc phn trong biu 
* lables: cung cp tn nhn cho mi phn biu .
* startangle mc nh l 0, s bt u phn u tinca biu  trn trc x dng. V d ny t startangle = 90 quay ngc chiu kim ng h 90  v phn u tin ca biu  a bt u trn trc y dng.
* autopc: dng  nh dng hin th nhn ca d liu, nu khng c hoc none th s hin th gi tr ca phn. nh dng % l %1.1f%%


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
names = ['Bob','Jessica','Mary','John','Mel']
absences = [3,0,1,0,8]
detentions = [2,1,0,0,1]
warnings = [2,1,5,1,2]
GradeList = zip(names,absences,detentions,warnings)
columns=['Names', 'Absences', 'Detentions','Warnings']
df = pd.DataFrame(data = GradeList, columns=columns)
df
```

```python
# V biu  cho trng d liu Warnings c nhn l trng Names, c gc quay 90 
plt.pie(df['Warnings'],labels=df['Names'],startangle=90, autopct='%1.1f%%',)
plt.axis('equal')
plt.show()
```

```python
# V biu  nh trn nhng phn biu  c nhn Mel lch ra so vi cc phn cn li v thm tiu  ca biu .
plt.pie(df['Warnings'],
labels=df['Names'],
explode=(0,0,0,0,0.15),
startangle=90,
autopct='%1.1f%%',)
plt.axis('equal')
plt.title('Biu  cho d liu cnh bo')
plt.show()
```
