# Bài 2: Nhập vào độ dài bán kính r của một hình tròn, in ra màn hình chu vi, diện tích của hình tròn đó. Yêu cầu các giá trị hiển thị trên màn hình có 2 chữ số sau dấu thập phân.

```python
# import library
from math import pi

# input r 
r = float(input("Nhập vào bán kính r = "))

# function calculate 
def perimeter(r):
    return (r*2)*pi

def area(r):
    return (r**2)*pi

# show result
print(f'Chu vi hình tròn là: %.2f ' % (perimeter(r)))
print(f'Diện tích hình tròn là:  %.2f ' % (area(r)))
```

# Bài 3: Nhập vào độ dài ba cạnh a, b, c của một tam giác. Tính diện tích của tam giác bằng công thức Heron như sau:  

```python
# import library 
import math

# input a, b, c
a = float(input("Nhập vào a = "))
b = float(input("Nhập vào b = "))
c = float(input("Nhập vào c = "))

# function
def heron(a, b, c): 
    p = (a+b+c)/2
    return math.sqrt(p*(p-a)*(p-b)*(p-c))

# show
print('Diện tích tam giác: ', heron(a, b, c))
```

# Bài 5: Nhập các hệ số, giải và biện luận các hệ phương trình sau theo quy tắc Cramer:


```python
# input
a1 = float(input("Nhập vào a1 = "))
b1 = float(input("Nhập vào b1 = "))
a2 = float(input("Nhập vào a2 = "))
b2 = float(input("Nhập vào b2 = "))
c1 = float(input("Nhập vào c1 = "))
c2 = float(input("Nhập vào c2 = "))

# function
def cramer(a1, b1, a2, b2, c1, c2):
    x = ((c1*b2) - (b1*c2))/((a1*b2 - b1*a2))
    y = ((a1*c2) - (c1*a2))/((a1*b2 - b1*a2))
    return x, y

# show
list = cramer(a1, b1, a2, b2, c1, c2)
print(f'Gía trị x = {list[0]}, y = {list[1]}')
```

# Bài 6: Viết chương trình giải phương trình bậc hai: ax2+bx+c=0

# Bài 7: Nhập chỉ số điện kế tháng trước và tháng này. Hãy tính số kW tiêu thụ và tiền điện cho mỗi hộ dựa vào bảng giá sau (áp dụng từ ngày 20/3/2019):


```python
# input 
t1 = int(input("Nhập vào t1 = "))
t2 = int(input("Nhập vào t2 = "))

# function 
def electricity_bill(t1, t2):
    bill_1 = 0
    bill_2 = 0
    if 0 <= t1 <= 50 or 0 <= t2 <= 50: 
        bill_1 = t1*1.678
        bill_2 = t2*1.678
    elif 51 <= t1 <= 100 or 51 < t2 < 100:
        bill_1 = t1*1.734
        bill_2 = t2*1.734
    elif 101 <= t1 <= 200 or 101 < t2 < 200:
        bill_1 = t1*2.014
        bill_2 = t2*2.014
    elif 201 <= t1 <= 300 or 201 < t2 < 300:
        bill_1 = t1*2.536
        bill_2 = t2*2.536
    elif 301 <= t1 <= 400 or 301 < t2 < 400:
        bill_1 = t1*2,834
        bill_2 = t2*2,834
    elif t1 > 401 or t2 > 401:
        bill_1 = t1*2,927
        bill_2 = t2*2,927
    return bill_1, bill_2

# show 
print(electricity_bill(t1, t2))
```

# Bài 8: Nhập độ dài ba cạnh a, b, c. Cho biết a, b, c có phải là ba cạnh của một tam giác không? Nếu là ba cạnh của một tam giác thì cho biết đó làm tam giác gì: đều, cân hay tam giác thường? 

```python
# input a, b, c, d
a = float(input("Nhập vào a = "))
b = float(input("Nhập vào b = "))
c = float(input("Nhập vào c = "))

def triangle(a, b, c):
    if (a + b >= c) or (a + c >= b) or (c + b >= a):
        if a == b == c:
            print('Tam giác đều')
        elif (a == b) or (a == c) or (b == c):
            print('Tam giác cân ')
        else: print('Tam giác thường ')

# show 
print(triangle(a, b, c))
```

# Bài 9: Nhập một số nguyên dương n (nếu nhập n≤0 thì chương trình phải yêu cầu nhập lại). Sau đó tính các tổng sau:

### Tính thừa số nguyên số 

```python
def thuasont(n):
    st=""
    for i in range(2,n):
        dem=0
        while(n%i==0):
            dem=dem+1
            n=n/i
            if (dem):
                if (dem>1):
                    st=st+str(i)+"^"+str(dem)
                else: st=st+str(i)
                if(n>i): st=st+"*"
    return st
a=int(input("nhap a="))
print(thuasont(a))
```
